```python
#sdf['Order Date']=pd.to_datetime(sdf['Order Date'],format='%m/%d/%Y')

import pandas as pd

# Ensure column names are clean (strip spaces)
sdf.columns = sdf.columns.str.strip()

# Convert dates
sdf['Order Date'] = pd.to_datetime(sdf['Order Date'], format='%m/%d/%Y', errors='coerce')
sdf['Ship Date'] = pd.to_datetime(sdf['Ship Date'], format='%m/%d/%y', errors='coerce')

# Calculate days difference
sdf['Days_Difference'] = (sdf['Ship Date'] - sdf['Order Date']).dt.days
```

```python
sum_sales=sdf['Sales'].sum().round(3)
print(sum_sales)
sum_profit=sdf['Profit'].sum().round(3)
print(sum_profit)
sum_Quantity=sdf['Quantity'].sum().round(3)
print(sum_Quantity)
sum_Discount=sdf['Discount'].sum().round(3)
print(sum_Discount)
```

```python
data_pie = group_sale['Sales'].head(10)
l = group_sale.index[:10]
e = [0.1,0.1,0.1]  # first slice exploded
clrs = ['blue','orange','green']

plt.figure(figsize=(5, 5))
plt.pie(data_pie, labels=l, explode=e, startangle=90,
            wedgeprops={'linewidth': 1, 'edgecolor': 'black'}, colors=clrs, autopct='%1.1f%%')

plt.title("Sales vs Segment")
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 cities by Sales
data_pie1 = group_city['Sales'].head(10)
l = group_city.index[:len(data_pie1)]  # labels matching the top 10
e = [0.2] * len(data_pie1)             # explode all slices slightly
clrs = ['blue', 'orange', 'green', 'red', 'magenta', 
        'purple', 'cyan', 'yellow', 'grey', 'brown'][:len(data_pie1)]  # 10 colors

plt.figure(figsize=(10, 10))
plt.pie(data_pie1, labels=l, explode=e, startangle=90,
        wedgeprops={'linewidth': 1, 'edgecolor': 'black'},
        colors=clrs, autopct='%1.1f%%')

plt.title("Top 10 Cities by Sales")
plt.axis('equal')  # keeps pie chart circular
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 cities by Sales
data_pie1 = group_city['Sales'].tail(10)
l = group_city.index[:len(data_pie1)]  # labels matching the top 10
e = [0.2] * len(data_pie1)             # explode all slices slightly
clrs = ['blue', 'orange', 'green', 'red', 'magenta', 
        'purple', 'cyan', 'yellow', 'grey', 'brown'][:len(data_pie1)]  # 10 colors

plt.figure(figsize=(6,6))
plt.pie(data_pie1, labels=l, explode=e, startangle=90,
        wedgeprops={'linewidth': 1, 'edgecolor': 'black'},
        colors=clrs, autopct='%1.1f%%')

plt.title("Top 10 Cities by Sales")
plt.axis('equal')  # keeps pie chart circular
plt.show()
```
